```python
df = df.drop(columns= 'rating')
```

```python
new_col_names = []
for i in df.columns:
  new_col_names.append(i.capitalize())
```

```python
df.columns = new_col_names
df.columns
```

```python
df['show_id']= df['show_id'].apply(lambda x: x.split('s')[0])
```

```python
type(df['show_id'][0])
```

```python
df["show_id"] = df["show_id"].astype(int)
```

```python
df['date_added'] = df['date_added'].apply(lambda x: x.replace('/' , '-'))
```

```python
def is_even(n):
    return n % 2 == 0
```

```python
def reverse_string(s):
    return s[::-1]
```

```python
def is_palindrome(s):
    return s == s[::-1]
```

```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for i in s:
        if i in vowels:
            count += 1
    return count
```

```python
def sort_list(L):
    return sorted(L)
```

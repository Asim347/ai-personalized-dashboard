```python
def sort_list(L):
    for i in range(len(L)):
        min_index = i
        for j in range(i+1, len(L)):
            if L[j] < L[min_index]:
                min_index = j

        # swap
        L[i], L[min_index] = L[min_index], L[i]

    return L


L = [1,2,3,4,5,-1,9,2]
print(sort_list(L))
```

```python
def largest(L):
    m = L[0]
    for i in L:
        if i > m:
            m = i
    return m

print(largest([2,5,1,9,3]))
```

```python
def check_even(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(check_even(6))
```

```python
def myAdd(*args):
  sum = 0
  for i in range(len(args)):
    sum += args[i]
  return sum

print(myAdd(2,3,-4,8,7))
```

```python
def printVariables(**args):
  for x in args:
    print("Variable name is: ", x, "and value is: " ,args[x])

printVariables(a = 4, b = 3, c = 5)
```
